# Breast Cancer Diagnosis — Exploratory Data Analysis

This project explores the Breast Cancer dataset to understand the features 
that differentiate malignant and benign tumors using exploratory data analysis and visualization.

```python
df=pd.read_csv("/kaggle/input/datasets/guriya79/breast-cancer/Breast_cancer_Reseach (1).csv")
```

```python
df["diagnosis"].value_counts()
```

```python
sns.countplot(x="diagnosis", data=df)
plt.title("Diagnosis Distribution")
plt.show()
```

```python
df["diagnosis"]=df["diagnosis"].map({"M":1,"B":0})
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Feature Correlation")
plt.show()
```

```python
sns.histplot(df["radius_mean"], kde=True)
plt.title("Distribution of Radius Mean")
plt.show()
```

```python
sns.boxplot(x="diagnosis", y="radius_mean", data=df)
plt.show()
```

```python
sns.scatterplot(x="radius_mean", y="texture_mean", hue="diagnosis", data=df)
plt.show()
```

```python
sns.pairplot(df, hue="diagnosis")
```

## Key Insights

- Malignant tumors tend to have larger radius and perimeter values.
- Features such as **radius_mean**, **perimeter_mean**, and **area_mean** show strong correlation.
- Visualizations show a clear separation between malignant and benign tumors.

## Conclusion

This exploratory data analysis revealed important patterns in the dataset.  
Several features strongly correlate with tumor diagnosis and may be useful for building machine learning models for cancer classification.
